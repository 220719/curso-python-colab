## Parte teórica
Estruturas condicionais permitem que o programa tome decisões. Elas são usadas quando o resultado depende de uma condição.


## Primeiro if


```python
idade = int(input('Digite sua idade: '))

if idade >= 18:
    print('Maior de idade')
else:
    print('Menor de idade')
```

## if, elif e else


```python
nota = float(input('Digite a nota: '))

if nota >= 7:
    print('Aprovado')
elif nota >= 4:
    print('Recuperação')
else:
    print('Reprovado')
```

## Condições aninhadas


```python
usuario = input('Usuário: ')
senha = input('Senha: ')

if usuario == 'admin':
    if senha == '1234':
        print('Acesso permitido')
    else:
        print('Senha incorreta')
else:
    print('Usuário não encontrado')
```

## Sistema de aprovação


```python
nota = float(input('Nota final: '))
frequencia = float(input('Frequência (%): '))

if nota >= 6 and frequencia >= 75:
    print('Aluno aprovado')
elif nota >= 4 and frequencia >= 75:
    print('Aluno em recuperação')
else:
    print('Aluno reprovado')
```
