## Parte teórica
Listas são estruturas usadas para armazenar vários valores em uma única variável. São fundamentais para programação e análise de dados.


## Criando listas


```python
notas = [7.5, 8.0, 6.5, 9.0]
print(notas)
print('Primeira nota:', notas[0])
print('Última nota:', notas[-1])
```

## Métodos de listas


```python
nomes = []
nomes.append('Ana')
nomes.append('João')
nomes.append('Maria')
print(nomes)

nomes.remove('João')
print(nomes)

nomes.insert(1, 'Carlos')
print(nomes)

removido = nomes.pop()
print('Removido:', removido)
print(nomes)
```

## Loop com listas


```python
notas = [7, 8, 5, 9, 10]

soma = 0
for nota in notas:
    soma += nota

media = soma / len(notas)
print('Média:', media)
```

## Sistema de notas


```python
notas = []

for i in range(5):
    nota = float(input(f'Digite a nota {i+1}: '))
    notas.append(nota)

media = sum(notas) / len(notas)
print('Notas:', notas)
print('Média:', media)
print('Maior nota:', max(notas))
print('Menor nota:', min(notas))
```
