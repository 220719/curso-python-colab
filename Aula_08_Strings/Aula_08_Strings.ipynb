## Parte teórica
Strings são textos em Python. Manipular textos é essencial para entrada de dados, relatórios, automações e processamento de informações.


## Métodos básicos


```python
texto = '  Python é muito útil para ciência de dados  '

print(texto.upper())
print(texto.lower())
print(texto.strip())
print(texto.replace('Python', 'Programação'))
```

## split e join


```python
frase = 'Python é uma linguagem poderosa'
palavras = frase.split()
print(palavras)

nova_frase = '-'.join(palavras)
print(nova_frase)
```

## f-strings


```python
nome = 'Ana'
nota = 8.75

mensagem = f'A aluna {nome} obteve nota {nota:.2f}'
print(mensagem)
```

## Analisador de frases


```python
frase = input('Digite uma frase: ')

palavras = frase.split()
quantidade_palavras = len(palavras)
quantidade_caracteres = len(frase)

print('Frase em maiúsculas:', frase.upper())
print('Frase em minúsculas:', frase.lower())
print('Quantidade de palavras:', quantidade_palavras)
print('Quantidade de caracteres:', quantidade_caracteres)
```
