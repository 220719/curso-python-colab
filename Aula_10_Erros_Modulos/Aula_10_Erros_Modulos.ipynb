Programas reais precisam lidar com erros. O tratamento de erros evita que o programa pare inesperadamente. Módulos ajudam a organizar melhor o código.


## Erro comum


```python
# Execute e observe o erro caso digite texto em vez de número.
numero = int(input('Digite um número inteiro: '))
print('Número digitado:', numero)
```

## Tratando erro com try/except


```python
try:
    numero = int(input('Digite um número inteiro: '))
    print('Número digitado:', numero)
except ValueError:
    print('Erro: você precisa digitar um número inteiro.')
```

## Tratando divisão por zero


```python
try:
    a = float(input('Digite o numerador: '))
    b = float(input('Digite o denominador: '))
    resultado = a / b
    print('Resultado:', resultado)
except ZeroDivisionError:
    print('Erro: não é possível dividir por zero.')
except ValueError:
    print('Erro: digite apenas números.')
finally:
    print('Fim do programa.')
```

## Importando módulos


```python
import math
import random

print('Raiz quadrada de 25:', math.sqrt(25))
print('Valor de pi:', math.pi)
print('Número aleatório:', random.randint(1, 10))
```

## Sistema robusto com tratamento de erros


```python
def solicitar_nota():
    while True:
        try:
            nota = float(input('Digite uma nota entre 0 e 10: '))
            if 0 <= nota <= 10:
                return nota
            else:
                print('A nota precisa estar entre 0 e 10.')
        except ValueError:
            print('Erro: digite um número válido.')

notas = []
for i in range(3):
    notas.append(solicitar_nota())

media = sum(notas) / len(notas)
print('Notas:', notas)
print('Média:', media)
```
