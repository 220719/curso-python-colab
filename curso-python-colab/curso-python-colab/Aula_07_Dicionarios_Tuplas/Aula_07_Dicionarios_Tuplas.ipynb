## Parte teórica
Tuplas, dicionários e sets são estruturas importantes do Python. Elas permitem organizar dados de formas diferentes.


## Tuplas


```python
coordenada = (10, 20)
print(coordenada)
print('x:', coordenada[0])
print('y:', coordenada[1])
```

## Dicionários


```python
aluno = {
    'nome': 'Ana',
    'idade': 20,
    'curso': 'Python',
    'nota': 8.5
}

print(aluno)
print(aluno['nome'])
print(aluno['nota'])
```

## Alterando dicionários


```python
produto = {'nome': 'Caderno', 'preco': 25.90, 'estoque': 10}

produto['preco'] = 29.90
produto['categoria'] = 'Papelaria'

print(produto)
print(produto.keys())
print(produto.values())
print(produto.items())
```

## Sets


```python
numeros = [1, 2, 2, 3, 3, 3, 4]
unicos = set(numeros)
print(unicos)
```

## Cadastro de produtos


```python
produtos = []

for i in range(3):
    nome = input('Nome do produto: ')
    preco = float(input('Preço: '))
    estoque = int(input('Estoque: '))

    produto = {'nome': nome, 'preco': preco, 'estoque': estoque}
    produtos.append(produto)

for produto in produtos:
    print(produto)
```
