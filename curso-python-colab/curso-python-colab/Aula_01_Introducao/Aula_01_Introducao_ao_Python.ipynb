## Parte teórica
Python é uma linguagem de programação muito usada em ciência de dados, automação, inteligência artificial, análise científica e desenvolvimento de sistemas. No Google Colab, podemos escrever e executar código diretamente no navegador.

Conceitos principais:
- **Variável:** espaço para armazenar uma informação.
- **Tipo de dado:** define a natureza da informação, como texto, número inteiro ou número decimal.
- **Comentário:** texto explicativo ignorado pelo Python.


## Primeiro código em Python


```python
print('Olá, mundo!')
print('Este é meu primeiro notebook de Python no Google Colab.')
```

## Variáveis e tipos de dados


```python
nome = 'Anuar'
idade = 35
altura = 1.75
estudando_python = True

print(nome)
print(idade)
print(altura)
print(estudando_python)

print(type(nome))
print(type(idade))
print(type(altura))
print(type(estudando_python))
```

## Entrada de dados


```python
nome = input('Digite seu nome: ')
print('Bem-vindo(a),', nome)
```

## Calculadora simples


```python
numero1 = float(input('Digite o primeiro número: '))
numero2 = float(input('Digite o segundo número: '))

soma = numero1 + numero2
subtracao = numero1 - numero2
multiplicacao = numero1 * numero2
divisao = numero1 / numero2

print('Soma:', soma)
print('Subtração:', subtracao)
print('Multiplicação:', multiplicacao)
print('Divisão:', divisao)
```
