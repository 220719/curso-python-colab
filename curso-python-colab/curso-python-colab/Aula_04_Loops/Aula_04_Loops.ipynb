## Parte teórica
Loops permitem repetir comandos automaticamente. Eles são úteis para processar listas, gerar sequências, validar entradas e automatizar tarefas.


## Loop for


```python
for i in range(1, 11):
    print('Número:', i)
```

## Loop while


```python
contador = 1

while contador <= 5:
    print('Contador:', contador)
    contador += 1
```

## break e continue


```python
for numero in range(1, 11):
    if numero == 5:
        continue
    if numero == 9:
        break
    print(numero)
```

## Gerador de tabuadas


```python
numero = int(input('Digite um número: '))

for i in range(1, 11):
    print(f'{numero} x {i} = {numero * i}')
```
