## Parte teórica
Arquivos permitem salvar dados de forma permanente. Essa aula mostra como criar, ler e atualizar arquivos usando Python puro.


## Escrevendo arquivo


```python
with open('meu_arquivo.txt', 'w', encoding='utf-8') as arquivo:
    arquivo.write('Primeira linha do arquivo.
')
    arquivo.write('Python pode salvar informações em arquivos.
')

print('Arquivo criado com sucesso!')
```

## Lendo arquivo


```python
with open('meu_arquivo.txt', 'r', encoding='utf-8') as arquivo:
    conteudo = arquivo.read()

print(conteudo)
```

## Adicionando conteúdo


```python
with open('meu_arquivo.txt', 'a', encoding='utf-8') as arquivo:
    arquivo.write('Nova linha adicionada.
')

with open('meu_arquivo.txt', 'r', encoding='utf-8') as arquivo:
    print(arquivo.read())
```

## Lendo dataset de frases


```python
# No GitHub, o arquivo estará na pasta datasets.
# No Colab, ajuste o caminho caso necessário.

caminho = '../datasets/frases.txt'

with open(caminho, 'r', encoding='utf-8') as arquivo:
    linhas = arquivo.readlines()

for linha in linhas:
    print(linha.strip())
```

## Sistema de armazenamento de dados


```python
nome = input('Nome: ')
idade = input('Idade: ')
curso = input('Curso: ')

with open('cadastro.txt', 'a', encoding='utf-8') as arquivo:
    arquivo.write(f'{nome};{idade};{curso}
')

print('Cadastro salvo com sucesso!')
```
