## Parte teórica
Operadores permitem realizar cálculos, comparações e testes lógicos. São fundamentais para qualquer programa em Python.


## Operadores matemáticos


```python
a = 10
b = 3

print('Soma:', a + b)
print('Subtração:', a - b)
print('Multiplicação:', a * b)
print('Divisão:', a / b)
print('Divisão inteira:', a // b)
print('Resto da divisão:', a % b)
print('Potência:', a ** b)
```

## Operadores relacionais


```python
idade = 18

print(idade == 18)
print(idade != 18)
print(idade > 18)
print(idade >= 18)
print(idade < 18)
print(idade <= 18)
```

## Operadores lógicos


```python
nota = 8
frequencia = 75

aprovado = nota >= 6 and frequencia >= 75
print('Aprovado?', aprovado)

recuperacao = nota >= 4 and nota < 6
print('Recuperação?', recuperacao)
```

## Sistema de média escolar


```python
nota1 = float(input('Nota 1: '))
nota2 = float(input('Nota 2: '))
nota3 = float(input('Nota 3: '))

media = (nota1 + nota2 + nota3) / 3
print('Média final:', media)
print('Média suficiente?', media >= 6)
```
